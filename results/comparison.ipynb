```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')

results = pd.read_csv('/content/drive/My Drive/DSAN6500_Project/TILS_project/results/Results_AllModels.csv')

results.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


training_accuracy_columns = ['Model', 'Training Accuracy', 'Training Top-5 Accuracy']


training_accuracy = results.loc[:, training_accuracy_columns]

print(training_accuracy)



training_accuracy_melted = training_accuracy.melt(id_vars='Model', var_name='Metric', value_name='Accuracy')

print(training_accuracy_melted)


palette = {'Training Accuracy': '#FF4F79', 'Training Top-5 Accuracy': '#A11692'}



plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Accuracy', hue='Metric', data=training_accuracy_melted, palette=palette)

# Add title and labels
plt.title('Comparison of Training Top-1 and Top-5 Accuracy by Model')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1.1)
plt.savefig('/content/drive/My Drive/DSAN6500_Project/TILS_project/results/training_accuracy_plot.png', format='png', dpi=300)
# Display the plot
plt.show()
```
